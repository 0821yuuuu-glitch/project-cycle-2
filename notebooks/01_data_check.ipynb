```python
import pandas as pd
import numpy as np

# --- 1. 讀取資料 ---
# 注意：路徑使用 '../' 是因為程式檔在 notebooks/ 資料夾內，需跳出一層才能進入 data/ [cite: 139, 145]
raw_data_path = '../data/raw/YRBS_2007.csv'
df = pd.read_csv(raw_data_path)

# --- 2. 變數定義與選擇 ---
# 根據作業要求，我們選擇一個行為變數與一個連續變數 [cite: 26, 43]
# 範例選擇：CurrentAlcoholUse (行為) 與 HowMuchDoYouWeighWithoutShoesInKG (連續)
target_cols = ['CurrentAlcoholUse', 'HowMuchDoYouWeighWithoutShoesInKG']
df_selected = df[target_cols].copy()

# --- 3. 資料檢查 (Data Check) ---
print("原始選取資料筆數:", len(df_selected))
print("缺失值統計:\n", df_selected.isnull().sum()) # 檢查缺失值 [cite: 81, 91]

# 移除缺失值 [cite: 68]
df_cleaned = df_selected.dropna().copy()
print("移除缺失值後筆數:", len(df_cleaned))

# --- 4. 行為變數重編碼 (Recoding) ---
# 根據規則：CurrentAlcoholUse 成功(1) 為 code 2-7, 失敗(0) 為 code 1 [cite: 37]
def recode_alcohol(x):
    if x in [2, 3, 4, 5, 6, 7]:
        return 1  # Success
    elif x == 1:
        return 0  # Failure
    else:
        return np.nan

df_cleaned['Alcohol_Binary'] = df_cleaned['CurrentAlcoholUse'].apply(recode_alcohol)

# --- 5. 儲存處理後的資料 ---
# 存入 processed 資料夾，供後續 EDA 與 Inference 使用 
output_path = '../data/processed/yrbs_recoded.csv'
df_cleaned.to_csv(output_path, index=False)

print(f"✅ 資料處理完成！已存至: {output_path}")
print("最終樣本數 (Final sample size):", len(df_cleaned)) # 記錄最終樣本數 [cite: 70]
```
